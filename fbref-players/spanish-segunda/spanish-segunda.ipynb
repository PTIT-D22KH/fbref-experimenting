```python
quantitative_col = [col for col in df2.columns if df2[col].dtypes != 'object' and df2[col].dtypes != 'string']
quantitative_col = quantitative_col[1:]
quantitative_col
```

```python
df5 = fbref.read_player_season_stats('possession')
```

```python


def remove_outliers_IQR(data, columns):
    filtered_data = data.copy()
    column0 = columns[0]
    column1 = columns[1]
    if column1 != '':
        # column = column_tuple[1]  # Use the second element of the tuple
        
        # Remove NaN values from the column
        valid_data = filtered_data[column0][column1].dropna()
        
        Q1 = np.percentile(valid_data, 25)
        Q3 = np.percentile(valid_data, 75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Apply the filtering to the copy of the DataFrame
        filtered_data = filtered_data[(filtered_data[column0][column1] >= lower_bound)]
    else:
        valid_data = filtered_data[column0].dropna()
        
        Q1 = np.percentile(valid_data, 25)
        Q3 = np.percentile(valid_data, 75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Apply the filtering to the copy of the DataFrame
        filtered_data = filtered_data[(filtered_data[column0] >= lower_bound)]
    return filtered_data
```

```python
df3 = df2.copy()
for col in quantitative_col:
    df3 = remove_outliers_IQR(df3,  col)
    print(df3.shape)
```
